```python
import requests
```

```python
url = 'https://movie.douban.com/subject/26752088/comments?start=20&limit=20&sort=new_score&status=P&comments_only=1'

heads = {
'Host': 'movie.douban.com',
'Referer': 'https://movie.douban.com/subject/26752088/comments?start=20&limit=20&sort=new_score&status=P',
'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36',
'X-Requested-With': 'XMLHttpRequest'
}
ret = requests.get(url, headers=heads)
print(ret)
```
